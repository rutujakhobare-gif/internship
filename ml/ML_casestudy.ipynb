```python
# Q1: Create a DataFrame with the following e-commerce sales data:
# Columns: OrderID, CustomerID, Product, Category, Quantity, Price, OrderDate, Rating, Review
# Create at least 50 sample records with realistic data including dates from 2023-2024
# Display the first 10 and last 5 records using head() and tail()

import numpy as np 
import pandas as pd
review=["Good", "Average", "Excellent"]
Product=['laptop','mobile','headphone','keyboard','smartwatch','tablet','mouse','camera','printer','charger']*5
category=['electronics','accessories','wearable','accessories','electronics']*10
data = {'OrderID':range(1,51),
        'CustomerID':range(101,151),
        'Product':np.random.choice(Product,50) ,
        'Quantity':np.random.randint(1,5,50),
        'category':np.random.choice(category,50),
        'Price':np.random.randint(10000,50000,50),
        'OrderDate': pd.to_datetime(np.random.choice(pd.date_range("2023-01-01", "2024-12-31"), 50)),
        'Rating':np.random.randint(1,5,50),
        
        'reviews' : np.random.choice(review,50)

}
df = pd.DataFrame(data)
print(df)
```

```python
# Q3: Introduce 10% missing values randomly in Rating and Review columns
# Detect missing values and display the count of missing values per column
# Fill missing ratings with the median rating and missing reviews with "No review provided"
```

```python
df.loc[2,'Rating']=None

df.loc[22,'Rating']=None

df.loc[42,'Rating']=None
print("Null value of Rating ",df['Rating'])
print("Missing values count",df['Rating'].isnull().sum())
```

```python
df.loc[12,'reviews']=None
df.loc[32,'reviews']=None
print("Null value of review ",df['reviews'])

print("Missing values count",df['reviews'].isnull().sum())
```

```python
df['Rating'].fillna(df['Rating'].median(),inplace=True)
print(df)
```

```python
df['reviews'].fillna("No review provided",inplace=True)
print(df)
```

```python
# Q4: Create a new column 'TotalAmount' by multiplying Quantity and Price
# Add another column 'DiscountedPrice' that applies 10% discount on orders above $30000
# Use conditional selection to filter orders where TotalAmount > $35000
```

```python
df['TotalAmount']=df['Quantity']*df['Price']
df['DiscountedPrice'] = df['Price'].where(df['Price'] <= 30000, df['Price'] * 0.9)
filtered_df = df[df['TotalAmount'] > 35000]
print(df)
print(filtered_df)
```

```python
# Q5: Select only the columns: Product, Category, TotalAmount, and Rating
# Filter and display all orders from the 'Electronics' category with Rating >= 4

filtered_df = df.loc[(df['category'] == 'electronics') & (df['Rating'] >= 4),['Product', 'category', 'TotalAmount', 'Rating']]

print(filtered_df)

```

```python
# Q6: Use boolean indexing to find all orders where:
# - Quantity > 2 AND Price < 50 OR Rating == 5
# Display the count of such orders

filtered_orders = df[((df['Quantity'] > 2) & (df['Price'] < 50)) | (df['Rating'] == 5)]

print("Count of orders:", filtered_orders.shape[0])
```

```python
# Q7: Sort the entire DataFrame by OrderDate (descending) and then by TotalAmount (descending)
# Create a new column 'SalesRank' that ranks products by TotalAmount in descending order

df_sorted=df.sort_values(by=['OrderDate','TotalAmount'],ascending=[False,False])
df_sorted
```

```python
df_sorted['SalesRank'] = df_sorted['TotalAmount'].rank(method='dense',ascending=False)
print('SalesRank\n',df_sorted['SalesRank'])
```

```python
# Q8: Find the top 5 best-selling products based on total quantity sold
# Display the product name and total quantity in descending order


top_5_products = (
    df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(5)
)

print(top_5_products)
```

```python
# GROUPING & AGGREGATION (Statistical Analysis)

# Q9: Group the data by 'Category' and calculate:
# - Total sales (sum of TotalAmount)
# - Average rating
# - Number of orders
# Display the results sorted by total sales
category_summary = (
    df.groupby('Category').agg(
          TotalSales=('TotalAmount', 'sum'),
          AverageRating=('Rating', 'mean'),
          NumberOfOrders=('OrderID', 'count')
      ).sort_values(by='TotalSales', ascending=False)
)

print(category_summary)
```
