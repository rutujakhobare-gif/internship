```python
# Q1: Create a simple line plot of y = 2x + 3 for x values ranging from 0 to 10.
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
x= np.linspace(0,10) #100 points betwen 0 and 10
y= 2 * x + 3

plt.plot(x,y)
plt.title("simple line plot of y= 2x + 3")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```

```python
x=np.linspace(0,10,50)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,color='red',linestyle='--',marker='o',label='sin(x)')
plt.plot(x,y2,color='blue',linestyle='-',marker='x',label='cos(x)')

plt.title(" line plot")
plt.xlabel("x-values")
plt.ylabel("y-values")

plt.grid(True)
plt.legend()

plt.show()
        
```

```python
# Q3: Create a bar chart showing the sales of four products: A, B, C, and D with random values.


categories=['A','B','C','D']
values=[20,43,56,78]

plt.bar(categories,values,color='pink')

plt.title("sales of product")
plt.xlabel("categories")
plt.ylabel("values")

plt.show()

```

```python
# Q4: Generate a scatter plot with random x and y values between 0 and 1, and use different colors for points.


x=np.random.rand(50)
y=np.random.rand(50)
colors=np.random.rand(50)
sizes=1000*np.random.rand(50)

plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='viridis')

plt.title("scattered plot example")
plt.xlabel("x-values")
plt.ylabel("y-values")

plt.colorbar() #add color scale
plt.show()
```

```python
# Q5: Plot a histogram for 1000 random numbers generated from a normal distribution.

data=np.random.randn(1000)

plt.hist(data,bins=30,color='green',edgecolor='black',alpha=0.7)

plt.title("histogram example")
plt.xlabel("data range")
plt.ylabel("frequency")

plt.show()
```

```python
# Q6: Draw a pie chart representing the percentage of five fruits: Apple, Banana, Mango, Grapes, and Orange.

labels=['Apple','Banana','mango','grapes','orange']
sizes=[20,15,25,30,10]
colors=['red','yellow','brown','green','orange']

plt.pie(sizes,labels=labels,colors=colors, autopct='%1.1f%%', startangle=70, shadow=True)

plt.title("The percentage of fruits")
plt.show()


```

```python
# Q7: Create two subplots in one figure  one showing a sine wave and the other showing a cosine wave.


import numpy as np

x= np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(x,y1,'r--')
plt.title("sin(x)")

plt.subplot(1,2,2)
plt.plot(x,y2,'b-')
plt.title("cos(x)")

plt.suptitle("multiple plot example")
plt.tight_layout()
plt.show()
```

```python
# Q8: Customize a line plot by adding title, axis labels, grid lines, and saving it as 'line_plot.png'.

x=np.linspace(0,5,100)
y=np.linspace(0,10,100)

plt.plot(x,y,'m-')
plt.title("Save image as line plot ")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.savefig("line_plot.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
# Q9: Create a 2x2 grid of subplots showing different plots: line, bar, scatter, and histogram.

fig,axs=plt.subplots(2,2,figsize=(10,8))

#line plot
x=np.linspace(0,10,100)
axs[0,0].plot(x,np.sin(x),'r')
axs[0,0].set_title("sine wave")

#bar plot
categories=['A','B','c','D']
values=[10,20,30,40]
axs[0,1].bar(categories,values,color='teal')
axs[0,1].set_title("Bar chart")

#histogram
data=np.random.randn(1000)
axs[1,0].hist(data,bins=25,color='purple',edgecolor='black')
axs[1,0].set_title("histogram")

#scatter plot
x=np.random.rand(50)
y=np.random.rand(50)
axs[1,1].scatter(x,y,color='orange')
axs[1,1].set_title("scattered plot")

plt.suptitle("matpltlib dashboard example",fontsize=14)
plt.tight_layout()

plt.savefig("matplotlib dashboard.png",dpi=300,bbox_inches='tight')
plt.show()
```

```python
# Q10: Create a line plot for y = e^x for x between 0 and 5, and change the color, marker, and line style.


# values of x from 0 to 5
x = np.linspace(0, 5, 100)

# y = e^x
y = np.exp(x)
plt.plot(x, y, color='brown', marker='o', linestyle='--')
plt.title("simple line plot of y= e^x")
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.show()
```
