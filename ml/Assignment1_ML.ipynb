```python
# 3. Import a CSV file (use any sample dataset) into a DataFrame.
#     Display the number of rows and columns using shape, and print column names.

data=pd.read_csv(r"C:\Users\rutuj\Downloads\data.csv")
data.shape
print(data)
```

```python
# 4. Add a new column named "Total_Marks" that sums all three subject scores.
#     Then create another column "Average" by dividing Total_Marks by 3.


data = {'student_name':['rishi','tanu','saniya','isha','meera'],
           'maths':[68,58,64,55,59],
           'science':[70,50,60,55,45],
           'english':[66,69,58,55,50]}

df=pd.DataFrame(data)

df['Total']=df['maths']+df['science']+df['english']

df['Average']=(df['Total']/3)

print(df.head())
```

```python
# 5. Introduce a missing value (NaN) in the "Science" column and then:
#     (a) Identify missing values using isnull()
import numpy as np
df.loc[2,'science']=None
print("with missing data values\n",df)
df['science'].isnull()

#     (b) Replace the missing value with the column mean using fillna().
mean=df['science'].mean()
df['science']=df['science'].fillna(df['science'].mean())

print("\n",df)
```

```python
# 6. Filter and display all students whose "Average" score is greater than 60.

print(df[df['Average'] > 60])
```

```python
# 7. Sort the DataFrame by "Total_Marks" in descending order and show the top 3 performer

print(df.sort_values(by='Total',ascending=False))
df.head(3)

```

```python
# 8. Group students based on their "English" score range (like above or below 70)

df['english_group']=np.where(df['english']>60,
                             'above 60',
                             'below 60')


#     and find the mean of "Maths" and "Science" marks for each group.

result= df.groupby('english_group')[['maths','science']].mean()
print(result)
```

```python
# 9. Merge two DataFrames:
#     (a) One with "Student_Name" and "Total_Marks"
#     (b) Another with "Student_Name" and "Sports_Score"
#     Then create a new column "Overall_Score" = Total_Marks + Sports_Score.

```

```python
df1 = pd.DataFrame({
    'student_name':['rishi','tanu','saniya','isha','meera'],
    'Total_marks':[204,177,182,165,154]
       })
df2 = pd.DataFrame({
    'student_name':['rishi','tanu','saniya','isha','meera'],
    'Sports_score':[10,6,8,7,9]
       })

merge_df=pd.merge(df1,df2,on='student_name')

merge_df['Overall_Score'] = merge_df['Total_marks'] + merge_df['Sports_score']
print(merge_df)
```
